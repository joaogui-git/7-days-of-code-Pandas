## Importação das bibliotecas e captura dos dados do Github

```python
#Importação das bibliotecas
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python

# URL do repositório GitHub
repo_url = 'https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/tree/main/Dia_1-Importando_dados/Datasets/dados_emprestimos'

# Obter o conteúdo da página do repositório
response = requests.get(repo_url)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Encontrar todos os links para arquivos .csv
    csv_files = []
    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.endswith('.csv'):
            csv_url = f'https://raw.githubusercontent.com{href}'
            csv_url = csv_url.replace('/blob','')
            csv_files.append(csv_url)

    # Exibir os arquivos .csv encontrados
    print('Arquivos .csv encontrados:')
    for csv_file in csv_files:
        print(csv_file)
        
else:
    print(f'Erro ao acessar o repositório: {response.status_code}')
```

```python
# Criando o DF de empréstimos
emprestimos = pd.DataFrame()

#Populando o DF de empréstimos
for csv in csv_files:
    dfatual = pd.read_csv(csv)
    emprestimos = pd.concat([emprestimos,dfatual])
```

```python
#Importação da base de dados dos exemplares
url = 'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet?utm_medium=email&_hsenc=p2ANqtz--_9mgf0io__chvBkmyNqZxMGOlajpWzz2W97AtbytBN2XosjLITyovRqeApns2SYtpM4bf-rsM0Neju9SMK4QIaDWThHX3_miS23ScxQvku9zXAR4&_hsmi=270874190&utm_content=270874190&utm_source=hs_automation'

exemplares = pd.read_parquet(url)
```

```python
#Testando para validar se a importação foi feita corretamente
exemplares.head()
```

```python
#Concatenando os empréstimos com as bibliotecas para sabermos quais exemplares foram alugados
df_concatenado = pd.merge(emprestimos, exemplares, how='left', on='codigo_barras')
```
