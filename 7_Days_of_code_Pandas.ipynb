## Criação e aplicação de dicionario com as categorizações

```python
dict_cdu = {
    range(0,100): 'Generalidades. Ciência e conhecimento.',
    range(100,200): 'Filosofia e psicologia.',
    range(200,300): 'Religião.',
    range(300,400): 'Ciências sociais.',
    range(400,500): 'Classe vaga. Provisoriamente não ocupada.',
    range(500,600): 'Matemática e ciências naturais.',
    range(600,700): 'Ciências aplicadas.',
    range(700,800): 'Belas artes.',
    range(800,900): 'Linguagem. Língua. Linguística.',
    range(900,1000): 'Geografia. Biografia. História'
}

def categorizar(valor_a_categorizar, dicionario_categorizacao):
    for r, categoria in dicionario_categorizacao.items():
        if valor_a_categorizar in r:
            return categoria
    return "Valor não encontrado"

```

```python
# Aplicando a categorização
df_concatenado['categoria'] = df_concatenado['localizacao'].apply(lambda x: categorizar(x,dict_cdu))
```

```python
df_concatenado.tail()
```

```python
#Validação da aplicação
df_concatenado['categoria'].value_counts()
```

## Convertendo as colunas para o formato Datetime

```python
df_concatenado.head()
```

```python
df_concatenado[['data_renovacao','data_emprestimo','data_devolucao']] = df_concatenado[['data_renovacao','data_emprestimo','data_devolucao']].apply(pd.to_datetime, format='%Y/%m/%d %H:%M:%S.%f')
```

```python
df_concatenado.info()
```

## Criação das colunas referente ao ano, mes e hora para ajudar nas análises solicitadas



```python
df_concatenado['mes_emprestimo'] = df_concatenado['data_emprestimo'].dt.month
df_concatenado['ano_emprestimo'] = df_concatenado['data_emprestimo'].dt.year
df_concatenado['horario_emprestimo'] = df_concatenado['data_emprestimo'].dt.hour
```

```python
df_concatenado.head()
```

Análise do número de exemplares emprestados por ano

```python
df_concatenado.groupby(['ano_emprestimo','mes_emprestimo'])[['id_exemplar']].nunique()
```

```python
exemplares_por_ano = df_concatenado.groupby(['ano_emprestimo'])[['id_exemplar']].nunique()
```

```python
exemplares_por_ano
```

```python
exemplares_por_ano.plot(kind='line', title='Exemplares por ano', xlabel='Ano', ylabel='Quantidade de exemplares', figsize=(10,6), color='red')
```

## Quantidade de exemplares emprestados por mês

```python
exemplares_por_mes = df_concatenado.groupby(['mes_emprestimo'])[['id_exemplar']].nunique()
exemplares_por_mes
```

```python
emprestimos_por_mes = df_concatenado.groupby(['mes_emprestimo'])[['id_exemplar']].count()
emprestimos_por_mes.sort_values(by='id_exemplar')
```

Os meses com maior quantidades de empréstimos são o mês de **Março seguido por Agosto. **

E os meses com menor demanda, logo os com maior possibilidade de programar atividades de não atendimento ou mesmo férias são os meses de **Janeiro e Dezembro**

```python
emprestimos_por_mes.plot(kind='line', title='Empréstimos por mês', xlabel='Mês', ylabel='Quantidade de empréstimos', figsize=(10,6))
```

## Quantidade de empréstimos realizados por hora

```python
emprestimos_por_hora = df_concatenado.groupby(['horario_emprestimo'])[['id_exemplar']].count()
emprestimos_por_hora
```

```python
emprestimos_por_hora.plot(kind='barh', title='Empréstimos por hora', xlabel='Quantidade de empréstimos', ylabel='Hora', figsize=(10,6))
```


Os horários diáriso com maior disponibilidade para alocar as atividades de não atendimento ao usuário seriam de manhã das 07:00 ás 08:00, e de noite, de preferência a partir das 21:00
